```python
import datetime
import numpy as np
import tensorflow as tf
import os
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# import the relevant functions from cnn functions
from cnn_functions import prep_training_data, prep_test_data, show_logger_plot
from cnn_functions import train_model, test_model, train_aug_gen_model

# set the seed value
seed = 42

# set pythonhashseed to a fixed value
os.environ['PYTHONHASHSEED']=str(seed)

# set the python pseudorandom number generator to a fixed value
random.seed(seed)

# set the numpy and tf pseudorandom number generator to a fixed value
np.random.seed(seed)
tf.compat.v1.set_random_seed(seed)

# configure new session
session_conf = tf.compat.v1.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)
sess = tf.compat.v1.Session(graph=tf.compat.v1.get_default_graph(), config=session_conf)
tf.compat.v1.keras.backend.set_session(sess)
```

```python
# Prepare the data for training and testing
csv_folder = "image-info-csv/"
train_csv_fname = f"{csv_folder}train_refined.csv"
train_folder_name = "train"
test_csv_fname = f"{csv_folder}test_base.csv"
test_folder_name = "test"

# initialise rescale value
img_rescale = 1./255
validation_split = 0.15
reshaped_size=[150,150]

# Generate the data for training and testing
x_train, x_val, y_train, y_val, x_input_shape, output_layer = prep_training_data(train_csv_fname, train_folder_name
                                                                 , validation_split, img_rescale, reshaped_size)

x_test, y_test = prep_test_data(test_csv_fname, test_folder_name, img_rescale, reshaped_size)

# initialise the callbacks that will be used for the models
earlyStopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy', restore_best_weights=True, patience=5)
current_time = datetime.datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
```

# Augmented the data and test on untuned model

- the cells below contain the function for augmenting the data
- a range of values have been tested for augmenting the data, and the resulting list of values used are show below

```python
def augment_data(x_train, y_train, x_val, y_val, batch_size):
    """
    augments the training data by performing scaling and data augmentation to generate more data

    Parameters:
    x_train (np.array): the input training data for the cnn
    y_train (np.array): the output training data for the cnn
    x_val (np.array): the input validation data for the cnn
    y_val (np.array): the input validation data for the cnn


    Returns: 
    tuple: A tuple containing two `tf.keras.preprocessing.image.ImageDataGenerator` objects.
    The first object generates batches of augmented training data, the second object generates batches of 
    validation data.
    """

    datagen = ImageDataGenerator(
        rotation_range=45,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.4,
        horizontal_flip=False
        )
    
    datagen.fit(x_train)

    # Use the data generator to generate batches of augmented data
    train_augment_generator = datagen.flow(x_train, y_train, batch_size=batch_size)
    val_augment_generator = ImageDataGenerator().flow(x_val, y_val, batch_size=batch_size)

    return train_augment_generator, val_augment_generator
```

# Comparison of training of untuned model on unaugmented and augmented data

By referring the the accuracy and validation accuracy of the model trained on unaugmented data and augmented data, we can make the following observations:

Model trained on unaugmented data:
- the training accuracy is close to 1, whereas the validation accuracy is around 0.9
- this implies that the model has overfitted to the training data as it is able memorise the training data while the validation accuracy lags behind

Model trained on augmented data:
- both the training accuracy and validation accuracy are close to each other at around 0.9
- this implies that the model is able to generalise well to new data and it is not overfitting to the data

However, in the above 2 comparisons, we can see that image augmentation has not led to a significant improvement in training the model. Some reasons for this are listed below:
- The current base model architecture is not well suited to this task and we may need to tweak the model architecture and hyperparameters
- As the validation and test accuracy is quite high at close to 0.90, there might be some problems with the data whereby some of the images is noisy and data augmentation is unable to help to address these issues due to limitations. For example, some of the images are blurry and some watermarks are present in some the images which may not be fixed using image augmentation techniques and may require new data to train the model on.

As the image augmentation helps to reduce overfitting, we will be using the augmented images in our subsequent training. In the next section, we will be tuning the model architecture and the hyperparameters. In the event the accuracy is not very much improved, we may revisit data augmentation to see if anything can be fixed.

```python
def create_tuned_model(x_input_shape, output_layer):
    """
    creates a cnn model

    Parameters:
    x_input_shape (tuple): the shape of the input data
    output_layer (int): the number of neurons in the output layer

    Returns:
    keras model object.
    """
    initializer = tf.keras.initializers.GlorotUniform(seed=42)
    
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Conv2D(filters=24, 
         kernel_size=(3, 3), activation="relu", input_shape=(x_input_shape), kernel_initializer=initializer))
    model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)
        ,strides=None ,padding="valid"))
    
    # add 3 convolutional layers
    model.add(tf.keras.layers.Conv2D(filters=24, 
         kernel_size=(3, 3), activation="relu", kernel_initializer=initializer))
    model.add(tf.keras.layers.Conv2D(filters=24, 
         kernel_size=(3, 3), activation="relu", kernel_initializer=initializer))
    model.add(tf.keras.layers.Conv2D(filters=24, 
         kernel_size=(3, 3), activation="relu", kernel_initializer=initializer))    
    
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(units=128, activation="relu", kernel_initializer=initializer))    
    model.add(tf.keras.layers.Dense(units=output_layer, activation="softmax", kernel_initializer=initializer))    
    model.compile(loss="categorical_crossentropy", optimizer="adam", 
                    metrics=["accuracy"])
    return model
```

```python
# Initialise callbacks

# Create a csv logger callback and create a unique name for the logger file using datetime
aug_tuned_logger_name = f"logger/train_model/03-tuned_augmented_data_model-{current_time}.log"
csv_aug_tuned_logger = tf.keras.callbacks.CSVLogger(aug_tuned_logger_name, separator=",", append=False)

# Create a list of the loggers to fit in the model
tuned_aug_callbacks = [earlyStopping, csv_aug_tuned_logger]
```

```python
tuned_epochs = 100
tuned_batch_size = 32

# generate steps per epoch
num_of_samples = len(x_train)
tuned_steps_per_epoch = num_of_samples // tuned_batch_size

# augment the data
train_augment_generator, val_augment_generator = augment_data(x_train, y_train, x_val, y_val, tuned_batch_size)

# create the model
aug_tuned_data_model = create_tuned_model(x_input_shape, output_layer)
aug_tuned_data_history = train_aug_gen_model(aug_tuned_data_model, train_augment_generator, tuned_batch_size, tuned_epochs
                                       , tuned_aug_callbacks, val_augment_generator, tuned_steps_per_epoch)

# test the model
print("The accuracy of the untuned augmented data model on the test dataset is: ")
print(test_model(aug_tuned_data_model, x_test, y_test))
```
