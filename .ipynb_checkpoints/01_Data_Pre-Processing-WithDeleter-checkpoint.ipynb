```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# List of classifiers to use for hot encoding
fruitList = ["apple", "banana", "orange", "mixed"]

# Get the file paths for the folders containing the train images and test
img_fpath_train = f"{os.getcwd()}/train/"
img_fpath_test = f"{os.getcwd()}/test/"

# folder to save the csv files to
folder="image-info-csv/"

# List of methods
# Creates a dataframe containing the file names of all the images
def load_img_names(fpath):
    df = pd.DataFrame()
    with os.scandir(fpath) as imgs:
        fname = []
        for img in imgs:
            if not img.name.startswith('.') and img.is_file():
                fname.append(img.name)
        df["fname"] = pd.Series(fname)
    return df



# This function inputs are dataframe of image names and the classifier names for hot encoding
def encode_onehot(fruitList, df):
    
    # Convert the fruit list and filename in df to lowercase to reduce errors
    fruitList = [fruit.lower() for fruit in fruitList]
    df["fname"] = df["fname"].str.lower()
    
    # Create a dataframe of categories and concatenate it with the original df
    classifier_df = pd.DataFrame(0,index=df.index, columns=fruitList)
    df = pd.concat([df, classifier_df],axis=1)
    
    # iterate through the list of fruits and assign the fruit with a number
    for fruit in fruitList:
        # if list contains partial string, assign it with the value 1
        bool_mat = df["fname"].str.contains(fruit)
        df.loc[bool_mat, fruit] = 1
        
    return df
```

# Use the code cells below for the training images

```python
# Get pandas df containing the fnames and classifiers
train_fname = load_img_names(img_fpath_train)
train_df = encode_onehot(fruitList,train_fname)
train_df.head()
```

# The code cells below are for updating the working csv files

```python
update_csv_fname = "train_update.csv"
delete_csv_fname = "train_delete.csv"

# Try to read the csv files to update
try:
    update_df = pd.read_csv(f"{folder}{update_csv_fname}")
except FileNotFoundError:
    print(f"File '{update_csv_fname}' not found. Continuing with empty DataFrame.")
    update_df = pd.DataFrame()
    
try:
    del_df = pd.read_csv(f"{folder}{delete_csv_fname}")
except FileNotFoundError:
    print(f"File '{delete_csv_fname}' not found. Continuing with empty DataFrame.")
    del_df = pd.DataFrame()    
```

```python
# Loop through each row in the dataframe and plot the corresponding image
for index, row in update_df.iterrows():
    #print(row[0])
    img = mpimg.imread(f"{img_fpath_train}{row[0]}")

    # Plot the image
    plt.imshow(img)
    plt.title(row[0])
    plt.show()
```

```python
# Loop through each row in the dataframe and plot the corresponding image
for index, row in del_df.iterrows():
    #print(row[0])
    img = mpimg.imread(f"{img_fpath_train}{row[0]}")

    # Plot the image
    plt.imshow(img)
    plt.title(row[0])
    plt.show()
```

```python
# replace the entries in update csv
base_df = pd.read_csv(f"{folder}train_refined.csv")

for fname in update_df["fname"]:
    if fname in base_df["fname"].tolist():
        base_index = base_df.loc[base_df['fname'] == fname].index.tolist()
        update_index = update_df.loc[update_df['fname'] == fname].index.tolist()
        base_df.iloc[base_index,1:] = update_df.iloc[update_index,1:]
        # show changes that are made
        print(base_df.iloc[base_index,:])
        
# save updated file
base_df.to_csv(f"{folder}train_refined.csv",index=False)
```

```python
# replace the entries in delete csv
base_df = pd.read_csv(f"{folder}train_refined.csv")

for fname in del_df["fname"]:
    if fname in base_df["fname"].tolist():
        base_index = base_df.loc[base_df['fname'] == fname].index.tolist()
        base_df.drop(base_index, inplace=True)
        print(f"index: {base_index} dropped!")
        
# save updated file
base_df.to_csv(f"{folder}train_refined.csv",index=False)
```
