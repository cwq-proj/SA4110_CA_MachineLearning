```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_training_logger(filepath, figsize=(12,6)):
    for folderPath in glob.glob(f"{filepath}/*/"):
        folderName = folderPath[len(filepath)+1:-1].replace("-", " ")
        files = f"{folderPath}/*.log"
        all_df = []
        values = []
        for log in glob.glob(files):
            start_index = log.find('&') + 1
            end_index = log.find('&', start_index)
            value = log[start_index:end_index]
            df = pd.read_csv(log)
            values.append(value)
            all_df.append(df)
        fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=figsize)

        # Iterate through each dataframe
        for i, df in enumerate(all_df):
            # Plot the training accuracy in the first plot
            ax1.plot(df['epoch'], df['accuracy'], label=f'Training: {values[i]}')

            # Plot the validation accuracy in the second plot
            ax2.plot(df['epoch'], df['val_accuracy'], label=f'Validation: {values[i]}')


        # Set plot labels and title for the first plot (training accuracy)
        ax1.set_xlabel("Epochs")
        ax1.set_ylabel("Training Accuracy")
        ax1.set_title("Training Accuracy")

        # Set plot labels and title for the second plot (validation accuracy)
        ax2.set_xlabel("Epochs")
        ax2.set_ylabel("Validation Accuracy")
        ax2.set_title("Validation Accuracy")


        # Add a legend to each plot
        ax1.legend(title=f"{folderName}", loc='lower right')
        ax2.legend(title=f"{folderName}", loc='lower right')

        # Adjust the spacing between plots
        plt.tight_layout()

        # Display the plots
        plt.show()
```

```python
aug_log_dir = os.path.join(os.getcwd(),"logger","data_aug_params")
plot_training_logger(aug_log_dir)
```
